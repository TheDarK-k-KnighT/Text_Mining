# Document Categorization

## Binary categories

```python
# Reading from corpus
import nltk
from nltk.corpus import PlaintextCorpusReader

documents_directory_name_politics = "C:\\Users\\The Dark Knight\\Desktop\\Applications\\Code"\
                                    "\\Text_Mining\\Mini_Newsgroups\\talk.politics.misc" 
documents_directory_name_comps = "C:\\Users\\The Dark Knight\\Desktop\\Applications\\Code\\Text_Mining"\
                                 "\\Mini_Newsgroups\\comp.os.ms-windows.misc"
corpus_politics = PlaintextCorpusReader(documents_directory_name_politics, '.*')
corpus_comps = PlaintextCorpusReader(documents_directory_name_comps, '.*')

corpus_politics.fileids()
```

```python
corpus_politics.raw('179097')
```

```python
import re
from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

# Custom preprocessor to lemmatize and remove stop words
def custom_preprocessor(text):
    text = re.sub(r'\W+|\d+|_', ' ', text)                   # removing numbers and punctuations
    text = nltk.word_tokenize(text)                          # tokenizing
    text = [word for word in text if not word in stop_words] # removing stopwords
    text = [lemmatizer.lemmatize(word) for word in text]     # lemmatizing
    return text


politics_news_documents = [(custom_preprocessor(corpus_politics.raw(fileid)), 'politics')
                for fileid in corpus_politics.fileids()]
politics_news_documents[0]
```

```python
politics_news_documents[len(politics_news_documents)-1]
```

```python
comps_news_documents = [(custom_preprocessor(corpus_comps.raw(fileid)), 'comps')
                       for fileid in corpus_comps.fileids()]
comps_news_documents[0]
```

```python
documents = politics_news_documents + comps_news_documents

import random
random.seed(50)
random.shuffle(documents)

print("Total number of documents = " + str(len(documents)))
print('\n')
documents[0]
documents[1]
```

```python
all_words = [word for document in documents for word in document[0]]
all_words_frequency = nltk.FreqDist(all_words)     # Frequency distribution of all words

print(all_words_frequency.most_common(150))
print('\n')
print(all_words_frequency['president'])
```

```python
word_features = all_words
word_features
```

```python
# Function to check presence of feature words in a document
def find_features(document_words):
    words = set(document_words)
    features = {}
    for word in word_features:
        features[word] = (word in words)
    return features

feature_sets = [(find_features(words), category) for (words, category) in documents]
feature_sets[1:5]
```

```python
train_set = feature_sets[:160]
test_set = feature_sets[160:]

# Using Naive Bayes Classifier
from nltk.classify.naivebayes import NaiveBayesClassifier

classifier = NaiveBayesClassifier.train(train_set)
print('Accuracy is : ', nltk.classify.accuracy(classifier, test_set))
classifier.show_most_informative_features(25)
```

## Multiple categories

```python
documents_directory_name_politics = "C:\\Users\\The Dark Knight\\Desktop\\Applications\\Code"\
                                    "\\Text_Mining\\Mini_Newsgroups\\talk.politics.misc" 
documents_directory_name_comps = "C:\\Users\\The Dark Knight\\Desktop\\Applications\\Code\\Text_Mining"\
                                 "\\Mini_Newsgroups\\comp.os.ms-windows.misc"
documents_directory_name_religion = "C:\\Users\\The Dark Knight\\Desktop\\Applications\\Code\\Text_Mining\\"\
                                    "Mini_Newsgroups\\talk.religion.misc"

corpus_politics = PlaintextCorpusReader(documents_directory_name_politics, '.*')
corpus_comps = PlaintextCorpusReader(documents_directory_name_comps, '.*')
corpus_religion = PlaintextCorpusReader(documents_directory_name_religion, '.*')

politics_news_documents = [(custom_preprocessor(corpus_politics.raw(fileid)), 'politics')
                          for fileid in corpus_politics.fileids()]

comps_news_documents = [(custom_preprocessor(corpus_comps.raw(fileid)), 'comp')
                       for fileid in corpus_comps.fileids()]

religion_news_documents = [(custom_preprocessor(corpus_religion.raw(fileid)), 'religion')
                          for fileid in corpus_religion.fileids()]

corpus_politics.fileids()
```

```python
documents = politics_news_documents + comps_news_documents + religion_news_documents

random.seed(50)
random.shuffle(documents)

print("Total number of documents = " + str(len(documents)))
```

```python
all_words = [word for document in documents for word in document[0]]
all_words_frequency = nltk.FreqDist(all_words)

print(all_words_frequency.most_common(150))
print('\n')
print(all_words_frequency['comp'])
```

```python
word_features = all_words
feature_sets = [(find_features(words), category) for (words, category) in documents]

train_set = feature_sets[:200]
test_set = feature_sets[200:]

# Using Naive Bayes Classifier
classifier = NaiveBayesClassifier.train(train_set)
print('Accuracy is : ', nltk.classify.accuracy(classifier, test_set))
classifier.show_most_informative_features(25)
```

## Using SVM to build model

```python
from nltk.classify.scikitlearn import SklearnClassifier
from sklearn.svm import SVC
svc = SklearnClassifier(SVC())
svc.train(train_set)
print("SV classifier accuracy on train set:", nltk.classify.accuracy(svc, train_set))
print("SV classifier accuracy on test set:", nltk.classify.accuracy(svc, test_set))
```

```python
# Using Support Vector Classifier with different configuration
svc = SklearnClassifier(SVC(kernel = 'poly', gamma = 0.01, C = 0.6))
svc.train(train_set)
print("SV classifier accuracy on train set:", nltk.classify.accuracy(svc, train_set))
print("SV classifier accuracy on test set:", nltk.classify.accuracy(svc, test_set))
```

```python
# Using Support Vector Classifier with  yet another configuration
svc = SklearnClassifier(SVC(gamma = 0.001, C = 1.2))
svc.train(train_set)
print("SV classifier accuracy on train set:", nltk.classify.accuracy(svc, train_set))
print("SV classifier accuracy on test set:", nltk.classify.accuracy(svc, test_set))
```

## Using RF to build model

```python
from sklearn.ensemble import RandomForestClassifier
rfc = SklearnClassifier(RandomForestClassifier())
rfc.train(train_set)
print("RF classifier accuracy on train set:", nltk.classify.accuracy(rfc, train_set))
print("RF classifier accuracy on test set:", nltk.classify.accuracy(rfc, test_set))
```

```python
# Using Random Forest Classifier with another configuration
from sklearn.ensemble import RandomForestClassifier
rfc = SklearnClassifier(RandomForestClassifier(n_estimators  = 50, max_depth = 3))
rfc.train(train_set)
print("RF classifier accuracy on train set:", nltk.classify.accuracy(rfc, train_set))
print("RF classifier accuracy on test set:", nltk.classify.accuracy(rfc, test_set))
```
